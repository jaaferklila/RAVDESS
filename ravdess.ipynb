<h2 style="color:red">Télechargement de Train dataset</h2>

```python
import pandas as pd
Train="archive/Train/"
emotion=[]
path=[]
for i in os.listdir(Train) :
    for aud in os.listdir(Train+i):
        split=aud.split(".")[0].split("-")
        
        if (split[2]=="01"):
            emotion.append("neutral")
            path.append(Train+i+"/"+aud)
        if(split[2]=="02"):
            emotion.append("calm")
            path.append(Train+i+"/"+aud)
        if(split[2]=="03"):
            emotion.append("happy")
            path.append(Train+i+"/"+aud)
        if(split[2]=="04"):
            emotion.append("sad")
            path.append(Train+i+"/"+aud)
        if(split[2]=="05"):
            emotion.append("angry")
            path.append(Train+i+"/"+aud)
        if(split[2]=="06"):
            emotion.append("fearful")
            path.append(Train+i+"/"+aud)
        if(split[2]=="07"):
            emotion.append("disgust")
            path.append(Train+i+"/"+aud)
        if(split[2]=="08"):
            emotion.append("surprised")
            path.append(Train+i+"/"+aud)
#convertir la liste des emotion en un dataframe
emotion_dataframe=pd.DataFrame(emotion,columns=['emotion'])
#convertir la lise des path en un dataframe
path_dataframe=pd.DataFrame(path,columns=["path"])
#concatination de deux datafarme(emotion et path)
pd_tarin=pd.concat([emotion_dataframe,path_dataframe],axis = 1)
            
```

```python
def extract_mfcc(file):
    y,sr=librosa.load(file,duration=3)
    mfcc=np.mean(librosa.feature.mfcc(y=y,sr=sr,n_mfcc=40).T,axis=0)
    return mfcc
```

```python
extract_mfcc(pd_tarin["path"][1])
```

```python
x_mfcc=pd_tarin["path"].apply(lambda x:extract_mfcc(x))
```
